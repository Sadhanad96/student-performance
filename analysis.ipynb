```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("student-mat.csv", sep=';')

# Show first 5 rows
data.head()
```

```python
# Continuous Assessment = Average of G1 and G2
data["Continuous"] = (data["G1"] + data["G2"]) / 2

print("Average Continuous Assessment:", round(data["Continuous"].mean(),2))
print("Average Final Assessment:", round(data["G3"].mean(),2))
```

```python
correlation = data["Continuous"].corr(data["G3"])
print("Correlation between Continuous and Final:", round(correlation,3))
```

```python
plt.figure()
sns.scatterplot(x=data["Continuous"], y=data["G3"])
plt.xlabel("Continuous Assessment Score")
plt.ylabel("Final Exam Score")
plt.title("Continuous Assessment vs Final Exam Performance")
plt.show()
```
